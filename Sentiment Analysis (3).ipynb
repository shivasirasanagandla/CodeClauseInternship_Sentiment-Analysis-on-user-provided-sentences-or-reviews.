# loading the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob
import re
from wordcloud import WordCloud
from collections import Counter
```

# General EDA

```python
def general_eda():
    # Manual input of dataset
    text = input("Enter the text: ")
    print("\nInput Text:", text)
    
    # Preprocessing
    text = preprocess_text(text)
    print("\nPreprocessed Text:", text)
    
    sentiment = analyze_sentiment_textblob(text)
    
    # Displaying sentiment
    print(f"\nSentiment: {sentiment}")
    print("\nText Statistics:")
    analyze_text_statistics(text)
    explore_ngrams(text)
    generate_wordcloud(text, sentiment)
```

# Analysis of the Sentiment Column

```python
def analyze_sentiment_column():
    # Manual input of dataset
    text = input("Enter the text: ")
    print("\nInput Text:", text)
    
    # Preprocessing
    text = preprocess_text(text)
    print("\nPreprocessed Text:", text)
    
    sentiment = analyze_sentiment_textblob(text)
    
    # Displaying sentiment
    print(f"\nSentiment: {sentiment}")
```

# Text Data Preprocessing

```python
def preprocess_text(text):
    text = text.lower()  # Convert text to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = re.sub(r'\s+', ' ', text)  # Remove extra whitespaces
    # Add more preprocessing steps such as removing stopwords, stemming, etc. for better accuracy
    return text
```

# Analyzing Text Statistics

```python
def analyze_text_statistics(text):
    word_count = len(text.split())
    char_count = len(text)
    avg_word_length = char_count / (word_count + 1e-8)
    print(f"Word Count: {word_count}")
    print(f"Character Count: {char_count}")
    print(f"Average Word Length: {avg_word_length:.2f}")
```

# Ngram exploration

```python
def explore_ngrams(text):
    ngrams = text.split()
    unigrams = Counter(ngrams)
    bigrams = Counter(zip(ngrams, ngrams[1:]))
    trigrams = Counter(zip(ngrams, ngrams[1:], ngrams[2:]))
    print("\nTop 10 Unigrams:", unigrams.most_common(10))
    print("\nTop 10 Bigrams:", bigrams.most_common(10))
    print("\nTop 10 Trigrams:", trigrams.most_common(10))
```

# Exploring the selected_text column

```python
# Exploring the selected_text column
def explore_selected_text():
    # Manual input of dataset
    text = input("Enter the text: ")
    print("\nInput Text:", text)
    
    # Preprocessing
    text = preprocess_text(text)
    print("\nPreprocessed Text:", text)
    
    sentiment = analyze_sentiment_textblob(text)
    
    # Displaying sentiment
    print(f"\nSentiment: {sentiment}")
```

# Wordclouds

```python
def generate_wordcloud(text, sentiment):
    # Manual input of dataset
    # No need to input text again, already preprocessed
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(f'Wordcloud for {sentiment} Sentiment')
    plt.axis('off')
    plt.show()
```

# Sentiment Extraction using TextBlob

```python
def analyze_sentiment_textblob(text):
    blob = TextBlob(text)
    sentiment = blob.sentiment.polarity
    if sentiment > 0:
        return 'Positive'
    elif sentiment < 0:
        return 'Negative'
    else:
        return 'Neutral'
```

# Example usage

```python
general_eda()
analyze_sentiment_column()
explore_selected_text()
generate_wordcloud(input("Enter the text: "), 'Positive')  # Change sentiment 
text_input = input("Enter the text: ").strip()  # Get user input and remove leading/trailing whitespace
if text_input:  # Check if input is not empty
    generate_wordcloud(text_input, 'Positive')  # Change sentiment accordingly
else:
    print("Input text is empty. Please provide some text.")
```
